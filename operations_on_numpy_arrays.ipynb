```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,40)
x
```

```python
sinx = np.sin(x)
sinx
```

```python
plt.plot(x, sinx)
```

```python
cosx = np.cos(x)
```

```python
plt.plot(x, sinx)
plt.plot(x, cosx)
plt.legend(['sinx', 'cosx'])
```

```python
# modifying style
plt.plot(x, sinx, 'x')
plt.plot(x, cosx, 'o')
plt.legend(['sinx', 'cosx'])
```

```python
diff = sinx**2 - cosx**2
plt.plot(x,diff)
```

```python
# dot treats arrays as vectors
np.dot(sinx,cosx)
```
